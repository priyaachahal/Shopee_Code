```python
import pandas as pd
import datetime
from datetime import datetime, timedelta
from collections import defaultdict

import collections

date_format = "%Y-%m-%d %H:%M:%S"
```

```python
df = pd.read_csv('./input_data/order_brush_order.csv')
df.head()
```

```python
# create a new column with 1 hour ahead
df['one_hour'] = df['event_time'].apply(lambda x: format(datetime.strptime(x, date_format)  + timedelta(hours=1),'%Y-%m-%d %H:%M:%S'))
df.head()
```

```python
# Sort the time based on event_time
df.sort_values('event_time',inplace=True)
```

```python
df.head()
```

```python
# groupby the dataframe into a list by shopid
df1 = df.groupby('shopid').agg({ 'orderid': lambda x: list(x),
                                'userid': lambda x: list(x),
                                'event_time': lambda x: list(x),
                               'one_hour': lambda x: list(x)})

df1.head(1)
```

```python
%%time

final_df = pd.DataFrame()
for index,row in df1.iterrows():
    if len(row['event_time'])>2:
        orders_dict = get_user_count(index,row)
        partial_df = pd.DataFrame(orders_dict).T
        final_df = final_df.append(partial_df)
        print(final_df.shape)
    else:
        orders_dict = {index:[{0:0}]}
        
        final_df = final_df.append(pd.DataFrame(orders_dict).T)
        print(final_df.shape)
```

```python
final_df.head()
final_df.reset_index(inplace=True)
final_df.columns=['shopid','users']
```

```python
users_list = []

for _,row in final_df.iterrows():

    if (any(x>=3 for x in row['users'].keys())):
        ind = [list(v) for k,v in row['users'].items() if k==max(row['users'].keys()) and max(row['users'].keys())>=3][0]

        if len(ind)>1:
            ind = ["&".join(str(i) for i in ind)][0]
            users_list.append(str(ind))
        elif len(ind) ==1:
            users_list.append(str(ind[0]))
    else:
        users_list.append("0")
```

```python
final_df['userid'] = users_list
order_brushing = final_df[['shopid','userid']]
order_brushing.head()
```
